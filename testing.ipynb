```python
import os
from dotenv import load_dotenv
from e2b_code_interpreter import Sandbox

load_dotenv()
api_key = os.environ["E2B_API"]

def execute_code(string):
    with Sandbox.create(api_key=api_key) as sandbox:
        execution = sandbox.run_code(string)
        print("stdout:", execution.logs.stdout)
        print("stderr:", execution.logs.stderr)
```
